# Example: Vertical 1D column simulation

## Setup

The setup is to build a 1-dimensional vertical column to investigate dust growth in the vertical direction. You initiate one radial zone and several vertical zone to build a vertical column of particles. This is adapted in the `setup.par` file along with the minimum and maximum radius of the simulation where `maxr-minr` must be sufficiently small enough. The next step is uncomment the `#VERTICALMODEL` flag in `preprocs.opt`.
An example for the setup of the vertical model can be accessed in `/setups/verticalcolumn/` directory. 

Note: Radial drift is turned off in this simulation to keep the vertical column mass constant (`RADRIFT` is turned off).

## Data visualisation

We now visualise an example simulation of a vertical column run using the `/setups/verticalcolumn` parameters.

```python
from mcdust import Simulation, Plots
```

```python
sim = Simulation()
```

```python
path='/your/path/to/data/directory'
sim.read(path)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from scipy.stats import kde
from scipy.constants import golden
from astropy import units as u
from astropy import constants as c
import os as os
au = u.au.to(u.cm)
k_b = c.k_B.cgs.value
m_p = c.m_p.cgs.value
Grav = c.G.cgs.value
Msun = c.M_sun.cgs.value
a0=1.e-4
amax= np.max(sim.swarms.grain_size)
```

```python
time = 1000. # years

it = sim.swarms.snapt.searchsorted(time)
dpi=300.
width = 6.
f,ax=plt.subplots(figsize=(width/2., width/golden),dpi=dpi)
ax.set_xscale("log")
ax.set_xlim(a0,amax)
ax.set_ylim(0,0.12)
im=ax.scatter(sim.swarms.grain_size[it,:],abs(sim.swarms.zdis[it,:]),c=sim.swarms.St[it,:],edgecolors='k',norm=LogNorm())
cbar=f.colorbar(im)
cbar.ax.set_ylabel("Stokes number")
ax.set_xlabel("grain size [cm]")
ax.set_ylabel("height above the midplane [au]")

ax.set_title('t = {:.1f} years'.format(sim.swarms.snapt[it]))#the argument step denotes the increment to print every nth variable. The default is 1.
```
